**Vector $\vec{v}$ has a value of [1, 2] but then a transformation happens. $\hat{j}$ lands at [2, 0] and $\hat{i}$ lands at [0, 1.5]. Where does $\vec{v}$ land?**

```python
v = np.array([1,2])
i_hat = np.array([2, 0])
j_hat = np.array([0, 1.5])

combined = np.array([i_hat, j_hat])

w = combined.dot(v)

w
```

**Vector $\vec{v}$ has a value of [1, 2] but then a transformation happens. $\hat{i}$ lands at [2, 1] and $\hat{j}$ lands at [1, -2]. Where does $\vec{v}$ land?**

```python
v = np.array([1,2])
i_hat = np.array([-2, 1])
j_hat = np.array([1, -2])

combined = np.array([i_hat, j_hat])

w = combined.dot(v)

w
```

**A transformation $\hat{i}$ lands at [1,0] and $\hat{j}$ lands at [2,2]. What is the determinant of this transformation?**

```python
i_hat = np.array([1, 0])
j_hat = np.array([2, 2])

combined = np.array([i_hat, j_hat])

det = np.linalg.det(combined)
det
```
