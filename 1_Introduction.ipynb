**Assuming no payments are made, how much would a $1,000 loan be worth at 5% interest compounded monthly after 3 years?**

```python
def calculate(value, period, percent):
    for _ in range(period):
        value += value * percent / 100
    return value
```

```python
print("The value after 2 periods is: " + str(round(calculate(1000, 3, 5), 2)))
```

**Assuming no payments are made, how much would a $1,000 loan be worth at 5% interest compounded monthly after 3 years?**

```python
def calculate(value, period, percent, compound_monthly=False):
    for _ in range(period):
        if compound_monthly:
            value += value * (percent/12) / 100
        else:
            value += value * percent / 100
    return value
```

```python
print("The value after 2 periods is: " + str(round(calculate(1000, 36, 5, True), 2)))
```

**For the function f(x) = 3x^2+1 whwat is the area under the curve for x between 0 and 2?**

```python
x = sp.symbols("x")
f = 3*x**2+1
m_slope = sp.integrate(f, (x, 0, 2))

print("The slope of the curve is: " + str(m_slope))
```
